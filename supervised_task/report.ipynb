# Model4 - CrossEntropy with weights

```python
# Load the data
train_loader, validation_loader = load_15SceneData()
device = torch.device("mps" if torch.backends.mps.is_available() else "cpu")

# Create model
model4 = Model1()

# setting a path
save_path = "models/model4.pt"

# set weights
trainset = datasets.ImageFolder('data/train', transform=transforms.ToTensor())
classes = trainset.classes
weights = []
for i in classes:
    if i in ["opencountry", "coast", "mountain"]:
        weights.append(1.5)
    else:
        weights.append(1)
# set criterion
criterion = torch.nn.CrossEntropyLoss(weight=torch.tensor(weights).to(device))

# training model
train_loss_model4, train_accuracy_model4, val_loss_model4, val_accuracy_model4 =\
    train(model4, train_loader, validation_loader, save_path, num_epochs=Epochs, learning_rate=LR, early_stopping_patience=10, criterion=criterion)

# plot loss
plot_loss(train_loss_model4, val_loss_model4)

# plot accuracy
plot_accuracy(train_accuracy_model4, val_accuracy_model4)

# save accuracy results
np.savetxt("results/accuracy_model4.csv", 
           val_accuracy_model4,
           delimiter =", ", 
           fmt ='% s')
```

# Final Model

```python
# Load the data
train_loader, validation_loader = load_15SceneData(crop=False)
device = torch.device("mps" if torch.backends.mps.is_available() else "cpu")

# Create model
model_final = Model1()

# setting a path
save_path = "models/model_final.pt"

# set weights
trainset = datasets.ImageFolder('data/train', transform=transforms.ToTensor())
classes = trainset.classes
weights = []

for i in classes:
    if i in ["opencountry", "coast", "mountain"]:
        weights.append(1.5)
    else:
        weights.append(1)

# set criterion
criterion = torch.nn.CrossEntropyLoss(weight=torch.tensor(weights).to(device))

# training model
train_loss_final, train_accuracy_final, val_loss_final, val_accuracy_final =\
    train(model_final, train_loader, validation_loader, save_path, num_epochs=Epochs, learning_rate=LR, early_stopping_patience=10, criterion=criterion)

# plot loss
plot_loss(train_loss_final, val_loss_final)

# plot accuracy
plot_accuracy(train_accuracy_final, val_accuracy_final)

# save accuracy results
np.savetxt("results/accuracy_model_final.csv", 
           val_accuracy_final,
           delimiter =", ", 
           fmt ='% s')
```
